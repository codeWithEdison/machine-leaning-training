```python
import matplotlib.pyplot as plt
import numpy as np

# Genrate some random  data 
data = np.random.randn(1000)
# create the histogram
plt.hist(data, color ='skyblue', edgecolor = 'black')
plt.xlabel('values')
plt.ylabel('frequency')
plt.title('histogram example')
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# lood dataset from csv 
data = pd.read_csv('music.csv')
# create the histogram
plt.hist(data['age'], color ='skyblue', edgecolor = 'black')
plt.xlabel('age')
plt.ylabel('frequency')
plt.title('histrogram of age in music data')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn  as sn
from sklearn.datasets import load_breast_cancer

data = load_breast_cancer()
plt.figure(figsize=(7.2, 3.5))
sn.kdeplot(data.data[:,0], fill=True)
plt.xlabel("mean radius")
plt.ylabel("density")
plt.show()
```
