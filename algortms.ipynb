```python
from sklearn.datasets import load_diabetes
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression 
from sklearn.metrics import mean_squared_error, r2_score

# lood diabets data

diabetes = load_diabetes()
diabetes

# split the datasets inyo trainng and testing 

x_train, x_test, y_train, y_test = train_test_split(diabetes.data[:,2], 
       diabetes.target, test_size =0.2, random_state= 0)

# reshape the input data

x_train = x_train.reshape(-1,1)
x_test = x_test.reshape(-1,1)

# create a linear regression object

lr_model = LinearRegression()
# fit the modelk on the traing data
lr_model.fit(x_train, y_train)

# make orediction on the test data

y_pred = lr_model.predict(x_test)

# calculate the mean squared error
mse = mean_squared_error(y_test, y_pred)

# root squred error

r2 = r2_score(y_test, y_pred)


print('mean squared Error: ', mse)
print('coeffient of determination: ', r2)


```

```python
import  matplotlib.pyplot as plt
# plot trainng data
plt.scatter(x_train, y_train, color='gray')

# plot regiretion line
plt.plot(x_train, lr_model.predict(x_train), color = 'red',  linewidth=2)
plt.xlabel("mean blood pressure")
plt.ylabel("disease progression")
#show the plot
plt.show()
```
